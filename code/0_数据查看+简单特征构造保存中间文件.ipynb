```python
def count2idx(num):
    if 0 <= num <= 10:
        return 0
    elif 10 <= num <= 50:
        return 1
    elif 50 <= num <= 150:
        return 2
    elif 150 <= num <= 300:
        return 3
    elif num >= 300:
        return 4
    if num < 0:
#         print(num)
        return 0
```

```python
def precision_score_cch(predictions, ground_truths):
    
    predictions, ground_truths = np.array(predictions)-1, np.array(ground_truths)-1
    y_pred = predictions
    y_true = ground_truths

    w=[1,10,50,100,300]
    n = len(y_true)
    count_r = [0 for i in range(5)]
    count = [0 for i in range(5)]
    for i in range(n):
        count[y_true[i]] += 1
        if y_pred[i] == y_true[i]:
            count_r[y_pred[i]] += 1
    sum1 = sum(w[i]*count_r[i] for i in range(5))
    sum2 = sum(w[i]*count[i] for i in range(5))
    precision = sum1/sum2
    return precision

# precision_score_cch(np.array(df_weibo_train['predict']).astype('int')+1, np.array(df_weibo_train['label']).astype('int')+1)
```

## 文本LDA

```python
from gensim import corpora, models
import jieba.posseg as jp, jieba
import time

stop_words = set()
def load_stopwords():
    """
    """
    with open('/data/cch/hyr/weibo/var/stopword.txt', 'r', encoding='UTF-8') as f:
        for line in f.readlines():
            stop_words.add(line.strip())

load_stopwords()

def get_lda_model(texts, topic_number):
    """
    """
    words_list = []
    flags = ('n', 'nr', 'ns', 'nt', 'eng', 'v', 'd') 
    for text in texts:
        words = [w.word for w in jp.cut(text) if w.flag in flags and w.word not in stop_words]
        words_list.append(words)
    dictionary = corpora.Dictionary(words_list)
    corpus = [dictionary.doc2bow(words) for words in words_list]
    lda = models.ldamodel.LdaModel(corpus=corpus, id2word=dictionary, num_topics = topic_number)
    return lda, corpus
```

```python
start = time.time()
lda_news, news_corpus = get_lda_model(list(df_weibo['WeiboText']), 20)
time.time() - start
```
